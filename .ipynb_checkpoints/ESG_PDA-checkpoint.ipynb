# Predictive Modeling and Machine Learning

In this section, I implemented supervised machine learning techniques to **predict ESG Risk Level, ESG Risk Score, and Controversy Score** using features such as sector, risk dimensions, and company size.

- Used **Random Forest Classifier** and **Random Forest Regressor**
- Performed **data encoding, train-test split, fitting, and evaluation**.
- Visualized prediction results using scatter plots and confusion matrix.

Before beginning the predictive analysis, I installed and imported all the essential Python libraries:

- **matplotlib** and **seaborn**: For data visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Predicting ESG Risk Level (Classification)

I predicted the **ESG Risk Level** (Low, Medium, High) using a Random Forest Classifier.

- Label encoded categorical columns: Sector, ESG Risk Level, and Controversy Level
- Input features include environmental, social, governance risk scores and controversy metrics.
- Evaluated model with a classification report and confusion matrix.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
```

```python
data=df.copy()
le=LabelEncoder()
```

```python
data['Sector']=le.fit_transform(data['Sector'])
data['ESG Risk Level']=le.fit_transform(data['ESG Risk Level'])
data['Controversy Level']=le.fit_transform(data['Controversy Level'])
```

```python
X1=data[['Sector','Full Time Employees','Environment Risk Score','Governance Risk Score','Social Risk Score','Controversy Score','Controversy Level']]
y1=data['ESG Risk Level']

X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,test_size=0.3,random_state=42)

clf=RandomForestClassifier(n_estimators=100,random_state=42)
clf.fit(X1_train,y1_train)
y1_predict=clf.predict(X1_test)

print("Classification Report:")
print(classification_report(y1_test, y1_predict))
print("Confusion Matrix:")
sns.heatmap(confusion_matrix(y1_test, y1_predict), annot=True, fmt='d', cmap='Blues')
plt.show()
```

### Insight:
The classification model provides a reasonably good prediction of ESG risk categories with distinct class-wise precision and recall scores.

## Predicting Total ESG Risk Score (Regression)

In this step, I performed regression using a **Random Forest Regressor** to predict the **total ESG risk score** of a company.

- Input features include sector, employee count, risk scores, and controversy metrics.
- Evaluated using **Mean Squared Error (MSE)** and **R² Score**.
- Visualized predicted vs actual values using a scatter plot.

```python
X2= data[['Sector','Full Time Employees','Environment Risk Score','Governance Risk Score','Social Risk Score','Controversy Score','Controversy Level']]
y2=data['Total ESG Risk Score']

X2_train,X2_test,y2_train,y2_test= train_test_split(X2,y2,test_size=0.3,random_state=42)

rg=RandomForestRegressor(n_estimators=100, random_state=42)
rg.fit(X2_train,y2_train)
y2_predict=rg.predict(X2_test)

print("Mean Squared Error:", mean_squared_error(y2_test, y2_predict))
print("R2 Score:", r2_score(y2_test, y2_predict))

plt.figure(figsize=(6, 6))
sns.scatterplot(x=y2_test, y=y2_predict)
plt.plot([y2_test.min(), y2_test.max()], [y2_test.min(), y2_test.max()], 'r--')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted ESG Risk Score")
plt.grid(True)
plt.show()

```

### Insight:
The model achieved a high R² score, indicating strong alignment with actual ESG risk scores.

## Predicting Controversy Score (Regression)

Here, I attempted to predict the **Controversy Score** using ESG scores and company attributes.

- Input features include risk scores, company size, sector, and ESG risk level.
- The model was evaluated using MSE and R² score.
- Visualized results with an actual vs predicted scatter plot.

```python
X3 = data[['Sector', 'Full Time Employees', 'Environment Risk Score',
           'Governance Risk Score', 'Social Risk Score', 'ESG Risk Level']]
y3 = data['Controversy Score']
X3_train,X3_test,y3_train,y3_test= train_test_split(X3,y3,test_size=0.3,random_state=42)

rg=RandomForestRegressor(n_estimators=100,random_state=42)
rg.fit(X3_train,y3_train)
y3_predict=rg.predict(X3_test)

print("Mean Squared Error:", mean_squared_error(y3_test, y3_predict))
print("R2 Score:", r2_score(y3_test, y3_predict))

plt.figure(figsize=(6, 6))
sns.scatterplot(x=y3_test, y=y3_predict)
plt.plot([y3_test.min(), y3_test.max()], [y3_test.min(), y3_test.max()], 'r--')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Controversy Score")
plt.grid(True)
plt.show()
```

### Insight:
The model showed **low predictive accuracy**, suggesting that controversy scores may depend on other qualitative or external factors not captured in this dataset.

## ESG Risk Segmentation Using Hierarchical Clustering

In this section, I have used hierarchical clustering to segment companies based on their ESG risk scores.  
This segmentation helps:

- Identify **distinct ESG profiles** (low-risk, medium-risk, high-risk clusters)
- Compare ESG performance **across sectors**
- Detect **clusters of concern or excellence** within industries
- Support better ESG benchmarking and strategy planning

After forming clusters using Agglomerative Clustering, I analyzed their distribution by sector using a count plot.

### Step 1: Importing Required Libraries for Hierarchical Clustering

Before performing ESG segmentation using Hierarchical Clustering, I imported key libraries:

- **StandardScaler** to normalize the data before clustering.
- **AgglomerativeClustering** to perform hierarchical (bottom-up) clustering.
- **silhouette_score** to evaluate the quality of different cluster solutions.
- **dendrogram** and **linkage** to visualize cluster hierarchies and guide optimal cluster count.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
```

### Step 2: Preprocessing ESG Scores for Clustering

I selected only the numerical ESG-related features relevant for segmentation:

- Used: Total ESG Risk Score, Environment Risk Score, and Governance Risk Score
- These features are standardized using StandardScaler so that no variable dominates due to differing scales.

```python
X = df[["Total ESG Risk Score", "Environment Risk Score", "Governance Risk Score"]]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Step 3: Dendrogram and Optimal Cluster Count

To decide how many ESG clusters exist, I took two approaches:

- **Dendrogram**: A tree-like diagram from hierarchical clustering that shows the order and distances at which clusters merge. A large vertical gap helps identify a good cut-off for the number of clusters.
- **Silhouette Score**: Measures how close each point is to its cluster vs other clusters. A higher score means better-defined clusters. I tested for clusters from 2 to 10.

This step ensured that clustering is data-driven rather than arbitrary.

```python
plt.figure(figsize=(10, 6))
dendrogram(linkage(X_scaled, method='ward'))
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Companies")
plt.ylabel("Euclidean Distance")
plt.show()

silhouette_scores = []
for k in range(2, 11):
    model = AgglomerativeClustering(n_clusters=k, linkage='ward')
    labels = model.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Optimal Cluster Selection")
plt.show()
```

### Step 4: Applying Hierarchical Clustering

Using the optimal number of clusters, I applied Agglomerative Clustering to segment companies:

- Assigned each company to a cluster based on ESG score similarities.
- Added a Cluster column to the DataFrame for further analysis.

```python
n_clusters = 3
hc = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
df['Cluster'] = hc.fit_predict(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Environment Risk Score', y='Governance Risk Score',
                hue='Cluster', palette='Set2', s=60, edgecolor='black')
plt.title("Clusters of ESG Risk Profiles")
plt.xlabel("Environmental Risk Score")
plt.ylabel("Governance Risk Score")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

### Step 5: Cluster Distribution by Sector

This visualization shows how companies in different sectors are distributed across ESG clusters.

- Uses countplot with hue="Cluster" to break down each sector into clusters.
- Helps observe which sectors are more prone to ESG risks.

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df, x='Sector', hue='Cluster', palette='Set2')
plt.title("Sector-wise Distribution Across ESG Clusters")
plt.xticks(rotation=45)
plt.show()
```

### Insight:

Some sectors (like **Energy** or **Healthcare**) show a larger share in higher-risk clusters, while others (like **Technology** or **Real Estate**) are mostly grouped into lower-risk clusters. This highlights sectors that may need more attention in ESG strategy and compliance.
