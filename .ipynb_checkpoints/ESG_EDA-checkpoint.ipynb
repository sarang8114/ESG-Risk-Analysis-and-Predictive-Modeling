## Data Cleaning and Preprocessing

In this section, I cleaned and prepared the **S&P 500 ESG Risk Ratings** dataset for analysis.

- Handled missing values in key columns using appropriate imputation methods.
- Converted string-based numerical fields into proper numeric format
- Renamed inconsistent column names for uniformity
- Checked for duplicates and ensured structural consistency

### Step 1: Importing Required Libraries

Before beginning the analysis, I installed and imported all the essential Python libraries:

- **pandas**: For data manipulation and analysis  
- **numpy**: For numerical operations  
- **matplotlib** and **seaborn**: For data visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Loading the Dataset
I loaded the dataset using pd.read_csv() to begin analyzing the ESG metrics of S&P 500 companies. This dataset contains risk-related information like ESG scores, controversies, employee count, sector classification, and more.

```python
df=pd.read_csv(r"C:\Users\Admin\Downloads\DA-project2\SP 500 ESG Risk Ratings.csv")
df.head()
```

Using df.info(), I examined the structure of the datasetâ€”checking the number of entries, column names, data types, and identifying the presence of missing values.

```python
df.info()
```

### Step 4: Generating Summary Statistics

I used df.describe() to get a statistical overview of numerical columns such as mean, standard deviation, minimum, and maximum values to understand data distribution.

```python
df.describe()
```

### Step 5: Checking for Missing Values
With df.isnull().sum(), I counted the number of missing values in each column to assess the extent of missing data and plan the necessary cleaning.

```python
df.isnull().sum()
```

### Step 6: Dropping Rows with Critical Missing Values
I dropped rows that had missing values in key columns like Address, Sector, Industry, Description, and Full Time Employees, since there were only one null value per attribute. 

```python
df=df.dropna(subset=['Address','Sector','Industry','Description','Full Time Employees'])
```

```python
df.isnull().sum()
```

### Step 7: Filling Missing Risk Score Values Sector and Industry Wise
To fill the missing Total ESG Risk Score values, I used the mean score of the respective sector to maintain logical consistency. Further, I filled missing values in the Environment Risk Score, Governance Risk Score, and Social Risk Score using the mean of their respective industry. Any remaining nulls were filled using the overall column mean.

```python
df['Total ESG Risk score']=df.groupby('Sector')['Total ESG Risk score'].transform(
    lambda x: x.fillna(x.mean())
)

for col in ['Environment Risk Score','Governance Risk Score','Social Risk Score']:
    df[col]=df.groupby('Industry')[col].transform(
        lambda x:x.fillna(x.mean())
    )
    df[col]=df[col].fillna(df[col].mean()) 
df.isnull().sum()
```

### Step 9: Creating ESG Risk Percentile
I computed an ESG Risk Percentile column using the .rank(pct=True) method, which ranks each score as a percentile out of 100 and filled in the missing values in ESG Risk Percentile column.


```python
df['ESG Risk Percentile'] = df['Total ESG Risk score'].rank(pct=True) * 100
```

```python
df.isnull().sum()
```

### Step 10: Assigning ESG Risk Levels
Based on the Total ESG Risk Score, I categorized companies into Low, Medium, and High ESG Risk Levels using an apply() function.

```python
def get_esg_risk_level(score):
    if score < 20:
        return 'Low'
    elif 20 <= score < 30:
        return 'Medium'
    else:
        return 'High'

df['ESG Risk Level']=df['Total ESG Risk score'].apply(get_esg_risk_level)

df.isnull().sum()
```

### Step 11: Filling Missing Controversy Scores Sector-Wise
For missing Controversy Score values, I used the mean score of the respective sector, assuming sectoral patterns influence controversy risks.


```python
df['Controversy Score']=df.groupby('Sector')['Controversy Score'].transform(
    lambda x: x.fillna(x.mean())
)
df.isnull().sum()
```

```python
df.head(20)
```

### Step 12: Converting Full-Time Employees to Numeric
The Full Time Employees column originally contained string values with commas (e.g., "1,000"). I converted them to numeric format using pd.to_numeric() for proper analysis.

```python
df['Full Time Employees'] = pd.to_numeric(df['Full Time Employees'].str.replace(',', ''), errors='coerce')
```

### Step 13: Categorizing Controversy Levels
Using the numeric Controversy Score, I filled in the missing Controversy Level column values with four categories:

- **Low**
- **Moderate**
- **Significant**
- **High**

This helps in grouping companies based on controversy risk severity.

```python
def map_controversy_level(score):
    if pd.isnull(score):
        return None
    elif score < 1.5:
        return 'Low Controversy Level'
    elif score < 2.5:
        return 'Moderate Controversy Level'
    elif score < 3.5:
        return 'Significant Controversy Level'
    elif score >= 3.5:
        return 'High Controversy Level'

df['Controversy Level'] = df['Controversy Score'].apply(map_controversy_level)
```

```python
df.isnull().sum()
```

### Step 14: Checking for Duplicate Entries
I checked for any duplicate rows using df.duplicated().sum() to ensure the dataset was clean and didn't contain redundant records.

```python
df.duplicated().sum()
```

### Step 15: Renaming Columns for Consistency
I renamed the Total ESG Risk score column to Total ESG Risk Score using rename() for consistency in naming conventions across the dataset.

```python
df.rename(columns={'Total ESG Risk score':'Total ESG Risk Score'},inplace=True)
df.columns
```

# Data Analysis and Visualization

In this section, I performed exploratory data analysis (EDA) to extract meaningful insights from the cleaned **SP 500 ESG Risk Ratings** dataset.

- Visualized ESG Risk Scores across different sectors and risk dimensions.
- Compared company sizes (Full-Time Employees) with ESG performance.
- Analyzed relationships between ESG risk, controversy levels, and governance scores.
- Used bar plots, scatter plots, box plots, and heatmaps for effective interpretation.

These visual insights help identify trends, correlations, and sector-specific patterns related to ESG risks.

## 1. What is the average ESG Risk Score across sectors?

I used a barplot to compare average ESG Risk Scores by sector using grouped mean values.

```python
df1=df.groupby('Sector')['Total ESG Risk Score'].mean().sort_values(ascending=True).reset_index()

plt.figure(figsize=(15,6))
sns.barplot(data=df1,x='Sector',y='Total ESG Risk Score',palette='coolwarm',hue='Sector')
plt.title('Distribution of ESG Risk Scores by Sector')
plt.xlabel('Sector')
plt.ylabel('Average ESG Risk Score')
plt.xticks(rotation=45)
plt.show()
```

**Insight:** Real Estate, Technology and Consumer Cyclical sectors generally have lower ESG risk, while Energy sector shows highest risk score.


## 2. What is the distribution of ESG Risk Score within each sector?

A boxplot was used to visualize the spread, median, and outliers of ESG Risk Scores within each sector.



```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Sector', y='Total ESG Risk Score', palette='Set2', hue='Sector')
plt.xticks(rotation=45)
plt.title('Distribution of ESG Risk Scores by Sector')
plt.show()
```

**Insight:** Certain sectors like Energy, Technology, Utilities, etc. exhibit wide variability and presence of outliers in ESG performance.


## 3. Does company size (by employees) influence ESG Risk?

A scatterplot was used to check correlation between number of employees and ESG scores, grouped by sector.

```python
df2 = df.groupby('Sector')['Full Time Employees'].mean().round().reset_index()
df3=df.groupby('Sector')['Total ESG Risk Score'].mean().reset_index()

df4=df2.merge(df3)

plt.figure(figsize=(15,5))
sns.scatterplot(data=df4, x='Full Time Employees', y='Total ESG Risk Score', palette='Set3',hue='Sector',s=70,edgecolor='black')
plt.title('ESG Risk Score vs Company Size')
plt.xlabel('Full Time Employees')
plt.ylabel('Average Total ESG Risk Score')
plt.grid(True)
plt.show()
```

**Insight:** There is no clear linear correlation between company size and ESG risk across sectors. Both large and small companies can fall anywhere on the ESG risk spectrum.


## 4. How do top ESG-performing companies score across E, S, and G?

I plotted a grouped bar chart for the top 10 companies with the lowest ESG Risk Scores to compare their E, S, and G component scores.

```python
top_performers = df.sort_values(by='Total ESG Risk Score').head(10)

labels = top_performers['Symbol']
x = np.arange(len(labels)) 
width = 0.25

plt.figure(figsize=(12, 6))

plt.bar(x - width, top_performers['Governance Risk Score'], width, label='Governance',edgecolor='black',color='coral')
plt.bar(x, top_performers['Social Risk Score'], width, label='Social',edgecolor='black',color='lightskyblue')
plt.bar(x + width, top_performers['Environment Risk Score'], width, label='Environment',edgecolor='black',color='palegreen')

plt.xlabel('Company Symbol')
plt.ylabel('Risk Score')
plt.title('Environmental, Social, and Governance Scores of Top 10 ESG Performers')
plt.xticks(x, labels, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

**Insight:** Top ESG performers consistently have low scores across all three components â€” especially environmental risks.


## 5. What are the average Environmental, Social, and Governance risks by sector?

Subplots were used to display E, S, and G risk scores across all sectors for easier comparison.

```python
df5=df.groupby('Sector')['Environment Risk Score'].mean().reset_index()
df6=df.groupby('Sector')['Social Risk Score'].mean().reset_index()
df7=df5.merge(df6)
df8=df.groupby('Sector')['Governance Risk Score'].mean().reset_index()
df8= df7.merge(df8)
df8

plt.figure(figsize=(18, 6))

# Subplot 1: Environmental
plt.subplot(1, 3, 1)
plt.barh(df8['Sector'], df8['Environment Risk Score'], color='lightgreen')
plt.title('Avg Environmental Risk Score')
plt.xlabel('Score')
plt.ylabel('Sector')

# Subplot 2: Social
plt.subplot(1, 3, 2)
plt.barh(df8['Sector'], df8['Social Risk Score'], color='skyblue')
plt.title('Avg Social Risk Score')
plt.xlabel('Score')
plt.ylabel('') 

# Subplot 3: Governance
plt.subplot(1, 3, 3)
plt.barh(df8['Sector'], df8['Governance Risk Score'], color='maroon')
plt.title('Avg Governance Risk Score')
plt.xlabel('Score')
plt.ylabel('') 

plt.tight_layout()
plt.show()
```

**Insight:** Energy sector has the highest environmental risk, while Finance has more governance-related risks.


## 6. Is there a relationship between controversy and ESG Risk?

A scatterplot was created to visualize how controversy scores correlate with ESG risk scores across companies and sectors.

```python
plt.figure(figsize=(10, 35))

sns.scatterplot(
    data=df, 
    x='Controversy Score', 
    y='Total ESG Risk Score', 
    hue='Sector', 
    palette='Set3',
    s=50,
    edgecolor='black'
)

plt.title('Controversy Score vs ESG Risk Score (Company-wise)')
plt.xlabel('Controversy Score')
plt.ylabel('Total ESG Risk Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Insight:** In general, companies with higher controversy scores also tend to have higher ESG risk, but there are notable outliers.


## 7. What are the strongest correlations among numeric features?

A heatmap was used to detect correlations between numerical features using Pearson correlation.

```python
df4=df.corr(numeric_only=True)
sns.heatmap(data=df4,cmap='viridis',annot=True)
```

**Insight:** ESG component scores are moderately correlated, and controversy score aligns somewhat with governance risk.


## 8. How are ESG Risk Levels distributed across sectors?

A stacked bar chart shows how many companies in each sector fall into Low, Medium, or High ESG risk categories.

```python
plt.figure(figsize=(12, 6))

df.groupby('Sector')['ESG Risk Level'].value_counts().unstack().plot(kind='bar', stacked=True, colormap='Set2', figsize=(14,6), edgecolor='black')

plt.title('Count of Companies by ESG Risk Level in Each Sector')
plt.xlabel('Sector')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)
plt.legend(title='ESG Risk Level')
plt.tight_layout()
plt.show()
```

**Insight:** Most sectors have a higher concentration of companies in the medium ESG risk category.

##  9. Is strong governance linked to fewer controversies?

A scatterplot compares governance risk scores against controversy scores to examine potential inverse relationships.

```python
plt.figure(figsize=(15,6))
sns.scatterplot(data=df, x='Governance Risk Score', y='Controversy Score')
plt.title('Is Strong Governance Linked to Fewer Controversies?')
plt.xlabel('Governance Risk Score')
plt.ylabel('Controversy Score')
plt.grid(True)
plt.show()
```

**Insight:**  Some weak negative relationship appears â€” companies with better governance may face fewer controversies.
